# Analysis of PbSn Plated HTS Tape Data (Current vs Voltage) Before and After Winding
## Data taken at 77K rig at 148 Sidney Street. In this notebook, each sample has three cells. The first cell is the calculation of high temperature superconducting performance degradation, the second an interactive plot of the performance of the tape before winding, the third an interactive plot of the performance of the tape after a winding.

Note: Some of the plots have different sample names than the sample section name they are in. The data is correct. The sample was just misnamed in the meta-data.<br>
Data collection and notebook by Denisse Cordova Carrizales (dcordovacarrizales@cfs.energy).
